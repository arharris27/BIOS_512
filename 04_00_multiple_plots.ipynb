5. Create a facets plot where each facet represents one iris species. 
(You may have to resize the output again.)

```r
options(repr.plot.width=4, repr.plot.height=3)

p=ggplot(iris, aes(x=Petal.Width, y=Petal.Length, size=Sepal.Width)) + geom_point()

p
```

```r
options(repr.plot.width=4, repr.plot.height=3)

p=ggplot(iris, aes(x=Petal.Width, y=Petal.Length, size=Sepal.Width)) +
geom_point(alpha=0.25) +
facet_wrap(~Species, ncol=2)

#Facet wrap subsets your data based on the values in the column selected, in this case, species. 
#Facet grid creates a grid of charts using more than one column. 

p
```
